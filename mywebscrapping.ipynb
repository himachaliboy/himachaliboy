```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url="https://www.hubertiming.com/results/2017GPTR10K"
html=url=urlopen(url)
```

```python
print(html)
```

```python
soup=BeautifulSoup(html,'html.parser')
soup
```

```python
title=soup.title.text
print(title)
```

```python
#print out the text
text=soup.get_text(strip=True)
print(text)
```

```python
#take hyper link
soup.find_all('a')
```

```python
#<a href="url">click</a>
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
    #print(link.text)
```

```python
#print the first 10 row for sanity check
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
#type(row_td)
```

```python
str_cells= str(row_td)
cleantext = BeautifulSoup(str_cells,"html.parser").get_text()
print(cleantext)
```

```python
import re
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('.</?\w>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
#print(clean2)
print(list_rows)
#type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df=df.replace('\\n','',regex=True)
df=df.replace('\\r','',regex=True)
df=df.replace('\<td>','',regex=True)
df=df.replace('\</td>','',regex=True)
df.head(20)
```

```python
df1 = df[0].str.split(',',expand=True)
df1.head(10)
```

```python
col_labels = soup.find_all('th')
col_labels
```

```python
all_header =[]
col_str =str(col_labels)
cleantext2 = BeautifulSoup(col_str, "html.parser").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2=pd.DataFrame(all_header)
df2
```

```python
df3= df2[0].str.split(',', expand=True)
df3
```

```python
frames =[df3, df1]
df4 = pd.concat(frames)
df4.head()
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df6 = df5.dropna(axis=0)
```
